```python
def evaluate():
    """
    True to continue game; False if game ends
    """
    if board['count']<5:
        print('Keep going.')
        return True
    elif result():
        if board['count']%2==0:
            print('Player 2 Wins.')
        else:
            print('Player 1 Wins.')
        return False
    elif board['count']==9:
        print('Game Tied.')
        return False
    else:
        print('Think before you move.')
        return True
```

```python
board={}

def display():
    print('\n    |     |\n',board[1],' | ',board[2],' | ',board[3],'\n____|_____|____\n    |     |\n',board[4],' | ',board[5],' | ',board[6],'\n____|_____|____\n    |     |\n',board[7],' | ',board[8],' | ',board[9],'\n    |     |\n')

def check_cell(p):
    """
    True if cell is occupied; False if cell is empty.
    """
    if (board[p]=='X' or board[p]=='O'):
        return True
    else:
        return False
    
def take_input():
    pos=' '
    while (pos not in '1 2 3 4 5 6 7 8 9'.split()) or check_cell(int(pos)):
        pos=input("Enter an unmarked, valid position on the board to be marked : ")
    pos=int(pos)
    if board['count']%2==0:
        board[pos]='X'
    else:
        board[pos]='O'
    board['count']+=1
    
def result():
    """
    True if we have a result; False if not.
    """
    if(board[1]==board[2]==board[3]!=' '):
        return True
    elif(board[4]==board[5]==board[6]!=' '):
        return True
    elif(board[7]==board[8]==board[9]!=' '):
        return True
    elif(board[1]==board[4]==board[7]!=' '):
        return True
    elif(board[2]==board[5]==board[8]!=' '):
        return True
    elif(board[3]==board[6]==board[9]!=' '):
        return True
    elif(board[1]==board[5]==board[9]!=' '):
        return True
    elif(board[3]==board[5]==board[7]!=' '):
        return True
    else:
        return False

def initialize():
    global board
    board = {1:' ',2:' ',3:' ',4:' ',5:' ',6:' ',7:' ',8:' ',9:' ','count':0}
    print('\t\t\t\tTic Tac Toe\nPlayer 1 is X, Player 2 is O.')
    display()
    print('Let the game begin.')
    while evaluate():
        take_input()
        display()
    print('Well Played !!')
```
