BlackJack Game with 4 card decks
1 Player vs Computer Dealer
Player can Hit or Stand
Dealer stands on soft 17
No Insurance, 3:2 on BlackJack, 1:1 on Win, 0 on Loss

```python
from random import choice
from random import shuffle

def value(card):
    if card in range(1,11):
        return card
    elif card=='A':
        return 11
    else:
        return 10
```

```python
class Player(object):
    
    def __init__(self):
        self.wins=self.losses=self.bet=self.count=0
        self.bank=1500
        self.hand=[]
        self.hard=True
        
    def setbet(self,bet):
        if 0<bet<=self.bank:
            self.bank-=bet
            self.bet=bet
            return True
        else:
            return False
        
    def hit(self,card):
        self.hand.append(card)
        if card=='A':
            if self.count>10:
                self.count+=1
            else:
                self.count+=11
                self.hard=False
        elif self.hard==False and self.count+value(card)>21 :
            self.count = self.count + value(card) - 10
            self.hard=True
        else:
            self.count+=value(card)
        
    def show(self):
        print('Your Hand:',end=' ')
        for c in self.hand:
            print(c,end=', ')
        print()
```

```python
class Dealer(object):
    
    def __init__(self):
        self.hand=[]
        self.count=0
        self.hard=True
        
    def hit(self,card):
        self.hand.append(card)
        if card=='A':
            if self.count>10:
                self.count+=1
            else:
                self.count+=11
                self.hard=False
        elif self.hard==False and self.count+value(card)>21 :
            self.count = self.count + value(card) - 10
            self.hard=True
        else:
            self.count+=value(card)
        
    def show(self):
        print("\tDealer's Hand:", end=' ')
        for c in self.hand:
            print(c,end=', ')
        print()
        
    def chance(self, CardsObject, pc):
        self.show()
        while self.count<pc and self.count<21 :
            print('\tDealer chooses to hit.')
            self.hit(CardsObject.draw())
            print('\tCard Drawn: ',self.hand[-1])
            self.show()
        if self.count>21 :
            return False
        else:
            print('Dealer chooses to stand.')
            return True
```

```python
class Cards(object):

    def __init__(self):
        deck=[x for x in range(2,11)]
        deck.insert(0,'A')
        deck+=['J','Q','K']
        self.shoe=deck*16

    def reset(self):
        self.shoe=self.deck*16
        
    def draw(self):
        new=choice(self.shoe)
        self.shoe.pop(self.shoe.index(new))
        return new
        
```
